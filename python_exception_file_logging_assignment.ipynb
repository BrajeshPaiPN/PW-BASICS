```python
# THEORY QUESTIONS - PART 1
# Q1. What is the difference between interpreted and compiled languages?
# Interpreted languages are executed line by line by an interpreter, while compiled languages are translated
# into machine code before execution. Python is generally interpreted, whereas C/C++ are compiled.

# Q2. What is exception handling in Python?
# Exception handling is a way to detect and manage runtime errors so a program can continue or fail gracefully.

# Q3. What is the purpose of the finally block in exception handling?
# The finally block always executes whether an exception occurs or not. It is commonly used for cleanup,
# such as closing files or releasing resources.

# Q4. What is logging in Python?
# Logging is the process of recording program events, warnings, errors, and debug information during execution.

# Q5. What is the significance of the __del__ method in Python?
# __del__ is a destructor-like method that may run when an object is about to be destroyed. It is not guaranteed
# to run immediately at a predictable time, so it should not be relied on for critical cleanup.

# Q6. What is the difference between import and from ... import in Python?
# import module loads the module and its members are accessed using module.member.
# from module import member imports selected names directly into the current namespace.

# Q7. How can you handle multiple exceptions in Python?
# You can use multiple except blocks or catch multiple exception types in one block using a tuple.

# Q8. What is the purpose of the with statement when handling files in Python?
# The with statement uses a context manager to open a file and automatically close it after the block ends.

# Q9. What is the difference between multithreading and multiprocessing?
# Multithreading runs multiple threads within one process and shares memory. Multiprocessing runs separate
# processes with separate memory spaces and is better for CPU-bound work in Python.

# Q10. What are the advantages of using logging in a program?
# Logging helps in debugging, monitoring, auditing, and understanding application behavior without relying only on print().
```

```python
# THEORY QUESTIONS - PART 2
# Q11. What is memory management in Python?
# Memory management is the allocation, use, and release of memory for Python objects during program execution.

# Q12. What are the basic steps involved in exception handling in Python?
# The common steps are: place risky code in try, handle errors in except, optionally run code in else when no
# exception occurs, and use finally for cleanup.

# Q13. Why is memory management important in Python?
# It prevents memory leaks, improves performance, and ensures efficient use of system resources.

# Q14. What is the role of try and except in exception handling?
# try contains code that may raise an exception. except contains the code that handles the exception.

# Q15. How does Python's garbage collection system work?
# Python primarily uses reference counting and also has a cyclic garbage collector to remove unreachable
# objects involved in reference cycles.

# Q16. What is the purpose of the else block in exception handling?
# The else block runs only if no exception occurs in the try block.

# Q17. What are the common logging levels in Python?
# DEBUG, INFO, WARNING, ERROR, and CRITICAL.

# Q18. What is the difference between os.fork() and multiprocessing in Python?
# os.fork() creates a child process directly on Unix-like systems and is low-level. multiprocessing is a
# higher-level, cross-platform module for creating and managing processes.

# Q19. What is the importance of closing a file in Python?
# Closing a file flushes buffered data, frees system resources, and avoids file corruption or locking issues.

# Q20. What is the difference between file.read() and file.readline() in Python?
# file.read() reads the full content or a specified number of characters. file.readline() reads only one line.

# Q21. What is the logging module in Python used for?
# The logging module is used to generate structured log messages for debugging and monitoring.

# Q22. What is the os module in Python used for in file handling?
# The os module helps work with file paths, check existence, rename files, remove files, and interact with the OS.

# Q23. What are the challenges associated with memory management in Python?
# Challenges include memory leaks from lingering references, high memory usage, circular references,
# and performance overhead from object creation and garbage collection.

# Q24. How do you raise an exception manually in Python?
# Use the raise statement, for example: raise ValueError('Invalid value')

# Q25. Why is it important to use multithreading in certain applications?
# Multithreading improves responsiveness and is useful for I/O-bound tasks such as network calls,
# file operations, and user interface work.
```

```python
# 1. Open a file for writing and write a string to it
with open('sample_write.txt', 'w', encoding='utf-8') as file:
    file.write('Hello, this text was written to the file.')

print('Data written to sample_write.txt')
```

```python
# 3. Handle a case where the file does not exist
try:
    with open('missing_file.txt', 'r', encoding='utf-8') as file:
        print(file.read())
except FileNotFoundError:
    print('The file does not exist.')
```

```python
# 4. Read from one file and write its content to another file
source_file = 'sample_write.txt'
destination_file = 'copied_output.txt'

with open(source_file, 'r', encoding='utf-8') as src:
    content = src.read()

with open(destination_file, 'w', encoding='utf-8') as dst:
    dst.write(content)

print(f'Copied data from {source_file} to {destination_file}')
```

```python
# 5. Catch and handle division by zero
try:
    result = 10 / 0
    print(result)
except ZeroDivisionError:
    print('Cannot divide by zero.')
```

```python
# 6. Log an error message to a log file when division by zero occurs
import logging

logging.basicConfig(
    filename='errors.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

try:
    value = 25 / 0
except ZeroDivisionError:
    logging.error('Division by zero exception occurred.')
    print('Logged division by zero error to errors.log')
```

```python
# 7. Log information at different levels
import logging

logger = logging.getLogger('level_demo')
logger.setLevel(logging.INFO)

if not logger.handlers:
    handler = logging.FileHandler('application.log')
    handler.setFormatter(logging.Formatter('%(levelname)s - %(message)s'))
    logger.addHandler(handler)

logger.info('This is an informational message.')
logger.warning('This is a warning message.')
logger.error('This is an error message.')

print('Logged INFO, WARNING, and ERROR messages to application.log')
```

```python
# 11. Handle an error when accessing a dictionary key that does not exist
student = {'name': 'Brajesh', 'age': 20}

try:
    print(student['grade'])
except KeyError:
    print('The requested key does not exist in the dictionary.')
```

```python
# 12. Use multiple except blocks to handle different exceptions
data = {'numbers': [10, 20, 30]}

try:
    index = int('abc')
    print(data['values'][index])
except ValueError:
    print('ValueError: invalid integer conversion.')
except KeyError:
    print('KeyError: dictionary key not found.')
except IndexError:
    print('IndexError: list index out of range.')
```

```python
# 14. Log both informational and error messages
import logging

logger2 = logging.getLogger('mixed_logger')
logger2.setLevel(logging.INFO)

if not logger2.handlers:
    handler2 = logging.FileHandler('mixed.log')
    handler2.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))
    logger2.addHandler(handler2)

logger2.info('Program started successfully.')

try:
    result = 5 / 0
except ZeroDivisionError as error:
    logger2.error('An error occurred: %s', error)

print('Logged informational and error messages to mixed.log')
```

```python
# 18. Basic logging setup with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

rotating_logger = logging.getLogger('rotating_logger')
rotating_logger.setLevel(logging.INFO)

if not rotating_logger.handlers:
    rotating_handler = RotatingFileHandler('rotating.log', maxBytes=1024 * 1024, backupCount=3)
    rotating_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))
    rotating_logger.addHandler(rotating_handler)

rotating_logger.info('This message goes to a rotating log file.')
print('Configured rotating log file: rotating.log')
```

```python
# 22. Check if a file is empty before reading its contents
import os

file_name = 'sample_write.txt'
if os.path.exists(file_name) and os.path.getsize(file_name) == 0:
    print('The file is empty.')
elif os.path.exists(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        print(file.read())
else:
    print('The file does not exist.')
```

```python
# 23. Write to a log file when an error occurs during file handling
import logging

file_logger = logging.getLogger('file_error_logger')
file_logger.setLevel(logging.ERROR)

if not file_logger.handlers:
    error_handler = logging.FileHandler('file_errors.log')
    error_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))
    file_logger.addHandler(error_handler)

try:
    with open('not_available.txt', 'r', encoding='utf-8') as file:
        print(file.read())
except OSError as error:
    file_logger.error('File handling error: %s', error)
    print('File handling error logged to file_errors.log')
```
