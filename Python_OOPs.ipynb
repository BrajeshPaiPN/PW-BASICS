```python
# Q1. What is Object-Oriented Programming (OOP)?
# OOP is a programming paradigm that organizes software design around data (objects) rather than
# functions and logic. It models real-world entities as 'objects' that have:
#   - Attributes (data/properties)
#   - Methods (behaviors/functions)
# Core pillars of OOP: Encapsulation, Abstraction, Inheritance, Polymorphism
# Example: A 'Car' object has attributes like color, speed, and methods like drive(), brake()
print("OOP: Organizing code around objects that bundle data and behavior together.")
```

```python
# Q2. What is a class in OOP?
# A class is a blueprint or template for creating objects.
# It defines the structure (attributes) and behavior (methods) that its objects will have.
# A class does NOT occupy memory until an object is instantiated from it.
# Syntax: class ClassName: ...

class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

print("Car is a class — a blueprint for creating car objects.")
```

```python
# Q3. What is an object in OOP?
# An object is an instance of a class — a concrete entity created from the class blueprint.
# Each object has its own copy of attributes but shares the class's methods.
# Creating an object: obj = ClassName(arguments)

my_car = Car("Toyota", "Red")   # 'my_car' is an object of class Car
print(f"Object: brand={my_car.brand}, color={my_car.color}")
```

```python
# Q4. What is the difference between Abstraction and Encapsulation?
#
# ABSTRACTION:
#   - Hides implementation complexity; shows only essential features to the user.
#   - Achieved via abstract classes and interfaces.
#   - 'What' an object does (not HOW it does it).
#   - Example: You press the 'Start' button in a car — you don't need to know engine internals.
#
# ENCAPSULATION:
#   - Bundles data + methods together and restricts direct access to some components.
#   - Achieved via private/protected attributes and getter/setter methods.
#   - 'How' data is protected inside an object.
#   - Example: A bank account hides the balance; you can only access it via deposit/withdraw methods.
#
# Key difference: Abstraction = hiding complexity | Encapsulation = hiding data
print("Abstraction hides 'how', Encapsulation protects 'what'.")
```

```python
# Q5. What are dunder methods in Python?
# Dunder (double underscore) methods, also called magic methods or special methods,
# are built-in methods with names like __init__, __str__, __repr__, __add__, etc.
# They allow operator overloading and customizing class behavior.
# Python automatically calls them in specific scenarios.
#
# Common dunder methods:
#   __init__  : Constructor — called on object creation
#   __str__   : Called by str() and print()
#   __repr__  : Called by repr() — unambiguous representation
#   __add__   : Called when + is used (operator overloading)
#   __len__   : Called by len()
#   __del__   : Destructor — called when object is destroyed
#   __call__  : Makes an object callable like a function

class Demo:
    def __init__(self):
        print("__init__ called: Object created!")
    def __str__(self):
        return "Demo Object"

d = Demo()
print(d)
```

```python
# Q6. Explain the concept of Inheritance in OOP.
# Inheritance allows a class (child/derived) to acquire properties and methods
# of another class (parent/base). Promotes code reuse and hierarchy.
#
# Types of Inheritance:
#   1. Single       : One parent → One child
#   2. Multiple     : Multiple parents → One child
#   3. Multilevel   : Grandparent → Parent → Child
#   4. Hierarchical : One parent → Multiple children
#   5. Hybrid       : Combination of the above
#
# Syntax: class Child(Parent): ...

class Animal:
    def eat(self):
        print("Eating...")

class Dog(Animal):         # Dog inherits from Animal
    def bark(self):
        print("Barking!")

d = Dog()
d.eat()    # Inherited method
d.bark()   # Own method
```

```python
# Q7. What is Polymorphism in OOP?
# Polymorphism means 'many forms'. It allows the same interface (method/operator)
# to behave differently based on the object type.
#
# Types:
#   1. Compile-time (Method Overloading) — not natively supported in Python
#   2. Runtime (Method Overriding) — child class overrides parent method
#   3. Duck Typing — Python style; objects are used based on behavior, not type

class Cat:
    def sound(self):
        print("Meow")

class Cow:
    def sound(self):
        print("Moo")

for animal in [Cat(), Cow()]:
    animal.sound()   # Same method call, different behavior = Polymorphism
```

```python
# Q8. How is Encapsulation achieved in Python?
# Encapsulation is achieved using access modifiers (by naming convention):
#   Public    : attribute       → accessible everywhere
#   Protected : _attribute      → accessible within class and subclasses
#   Private   : __attribute     → accessible only within the class (name-mangled)
#
# Python uses getter and setter methods (or @property decorator) to access private data.

class Person:
    def __init__(self, name, age):
        self.name = name         # Public
        self._dob = '2000-01-01' # Protected
        self.__salary = 50000    # Private

    def get_salary(self):        # Getter
        return self.__salary

p = Person("Alice", 25)
print(p.name)
print(p.get_salary())
# print(p.__salary)  # Raises AttributeError
```

```python
# Q9. What is a Constructor in Python?
# A constructor is a special method called automatically when an object is created.
# In Python, the constructor is __init__(self, ...).
# Purpose: Initialize instance attributes when an object is instantiated.
#
# Types:
#   - Default Constructor   : __init__(self) — no parameters
#   - Parameterized Constructor: __init__(self, params) — with parameters

class Student:
    def __init__(self, name, roll):   # Constructor
        self.name = name
        self.roll = roll
        print(f"Student created: {self.name}, Roll: {self.roll}")

s = Student("Bob", 101)   # Constructor auto-called
```

```python
# Q10. What are Class methods and Static methods in Python?
#
# INSTANCE METHOD:
#   - Operates on instance attributes; takes 'self' as first argument
#
# CLASS METHOD (@classmethod):
#   - Operates on class-level data; takes 'cls' as first argument
#   - Decorated with @classmethod
#   - Can be called on class or instance
#
# STATIC METHOD (@staticmethod):
#   - Does NOT access class or instance data
#   - Utility method that belongs to class namespace
#   - Decorated with @staticmethod; no 'self' or 'cls'

class MyClass:
    class_var = "I am a class variable"

    def instance_method(self):
        print(f"Instance method called, class_var: {self.class_var}")

    @classmethod
    def class_method(cls):
        print(f"Class method called, class_var: {cls.class_var}")

    @staticmethod
    def static_method():
        print("Static method — no access to class or instance data")

obj = MyClass()
obj.instance_method()
MyClass.class_method()
MyClass.static_method()
```

```python
# Q11. What is Method Overloading in Python?
# Method Overloading = defining multiple methods with the same name but different parameters.
# Python does NOT support traditional method overloading (unlike Java/C++).
# In Python, if you define two methods with the same name, the second overwrites the first.
#
# Python achieves similar behavior using:
#   - Default arguments
#   - *args / **kwargs

class Calculator:
    def add(self, a, b=0, c=0):   # Simulating overloading with defaults
        return a + b + c

calc = Calculator()
print(calc.add(5))          # 5
print(calc.add(5, 3))       # 8
print(calc.add(5, 3, 2))    # 10
```

```python
# Q12. What is Method Overriding in OOP?
# Method Overriding occurs when a child class provides its own implementation
# of a method already defined in the parent class.
# The child method has the SAME name and signature as the parent method.
# Enables Runtime Polymorphism.

class Shape:
    def draw(self):
        print("Drawing a generic shape")

class Circle(Shape):
    def draw(self):              # Overrides parent's draw()
        print("Drawing a Circle")

s = Shape()
s.draw()
c = Circle()
c.draw()   # Child's version is called
```

```python
# Q13. What is a Property Decorator in Python?
# @property allows you to define a method that can be accessed like an attribute.
# It provides a way to add getter, setter, and deleter logic cleanly.
# Benefits: Controlled access to private data, validation, computed properties.

class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):                # Getter
        return self._celsius

    @celsius.setter
    def celsius(self, value):         # Setter with validation
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value

    @property
    def fahrenheit(self):             # Computed property
        return (self._celsius * 9/5) + 32

t = Temperature(25)
print(f"Celsius: {t.celsius}, Fahrenheit: {t.fahrenheit}")
t.celsius = 30
print(f"Updated Celsius: {t.celsius}")
```

```python
# Q14. Why is Polymorphism important in OOP?
# Polymorphism is important because:
#   1. Code Reusability: Write generic code that works with different object types.
#   2. Flexibility: New classes can be added without changing existing code.
#   3. Extensibility: Makes it easy to extend programs.
#   4. Simplifies code: Use a single interface for multiple implementations.
#   5. Supports Open/Closed Principle (SOLID): Open for extension, closed for modification.
#   6. Enables dynamic dispatch (runtime polymorphism) for flexible program behavior.
print("Polymorphism enables flexible, reusable, and extensible code.")
```

```python
# Q15. What is an Abstract Class in Python?
# An abstract class is a class that cannot be instantiated directly.
# It contains one or more abstract methods that MUST be implemented by subclasses.
# In Python, use the 'abc' module (Abstract Base Class).
# Purpose: Enforce a contract — all subclasses must implement specified methods.

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def fuel_type(self):
        pass   # Subclass MUST implement this

class Car(Vehicle):
    def fuel_type(self):
        return "Petrol"

# v = Vehicle()  # TypeError: Can't instantiate abstract class
c = Car()
print(c.fuel_type())
```

```python
# Q17. What is the difference between a class variable and an instance variable?
#
# CLASS VARIABLE:
#   - Defined inside the class body, outside any method.
#   - Shared across ALL instances of the class.
#   - Accessed via ClassName.variable or self.variable.
#   - Changes affect all instances.
#
# INSTANCE VARIABLE:
#   - Defined inside __init__ using 'self.variable'.
#   - Unique to each object (instance).
#   - Changes affect only that specific object.

class Employee:
    company = "PW Skills"          # Class variable (shared)

    def __init__(self, name):
        self.name = name           # Instance variable (unique per object)

e1 = Employee("Alice")
e2 = Employee("Bob")
print(e1.company, e1.name)
print(e2.company, e2.name)
Employee.company = "Updated PW"  # Changes for ALL instances
print(e1.company, e2.company)
```

```python
# Q18. What is Multiple Inheritance in Python?
# Multiple Inheritance allows a class to inherit from MORE THAN ONE parent class.
# Python supports this natively (unlike Java which uses interfaces instead).
# MRO (Method Resolution Order) — Python uses C3 linearization to resolve conflicts.
# Use ClassName.__mro__ to see the order.

class Father:
    def skills(self):
        print("Father: Coding")

class Mother:
    def skills(self):
        print("Mother: Painting")

class Child(Father, Mother):   # Multiple Inheritance
    pass

c = Child()
c.skills()                     # Father's method called (MRO)
print(Child.__mro__)           # Shows resolution order
```

```python
# Q19. Explain the purpose of __str__ and __repr__ methods in Python.
#
# __str__:
#   - Returns a human-readable (informal) string representation of the object.
#   - Called by str(obj) and print(obj).
#   - Meant for end users.
#
# __repr__:
#   - Returns an unambiguous, developer-friendly string representation.
#   - Called by repr(obj) and when object is shown in the interpreter.
#   - Ideally, eval(repr(obj)) should recreate the object.
#   - If __str__ is missing, Python falls back to __repr__.

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point({self.x}, {self.y})"         # User-friendly

    def __repr__(self):
        return f"Point(x={self.x}, y={self.y})"     # Dev-friendly

p = Point(3, 4)
print(str(p))   # Calls __str__
print(repr(p))  # Calls __repr__
```

```python
# Q20. What is the significance of the super() function in Python?
# super() returns a proxy object that delegates method calls to the parent class.
# Uses:
#   1. Call parent's __init__ to avoid code repetition.
#   2. Call overridden parent methods from child.
#   3. Works correctly in Multiple Inheritance with MRO.

class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal __init__ called for {self.name}")

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # Calls Animal's __init__
        self.breed = breed
        print(f"Dog __init__ called: breed={self.breed}")

d = Dog("Rex", "Labrador")
```

```python
# Q21. What is the significance of the __del__ method in Python?
# __del__ is the destructor method — called when an object is about to be garbage collected.
# Python uses reference counting for garbage collection.
# __del__ is called when the reference count drops to zero.
# Use cases: Releasing resources (file handles, DB connections, network sockets).
# Note: Not guaranteed to be called at a specific time; avoid relying on it for critical cleanup.
# Prefer context managers (with statement) for resource management.

class Resource:
    def __init__(self, name):
        self.name = name
        print(f"Resource '{self.name}' acquired.")

    def __del__(self):
        print(f"Resource '{self.name}' released (destructor called).")

r = Resource("DatabaseConnection")
del r   # Triggers __del__
```

```python
# Q22. What is the difference between @staticmethod and @classmethod in Python?
#
# @staticmethod:
#   - No access to 'self' (instance) or 'cls' (class).
#   - Pure utility function attached to a class for organization.
#   - Called: ClassName.method() or instance.method()
#
# @classmethod:
#   - Receives 'cls' as the first argument (the class itself).
#   - Can access/modify class-level attributes.
#   - Often used as alternative constructors (factory methods).
#   - Called: ClassName.method() or instance.method()

class MyClass:
    count = 0

    @classmethod
    def increment(cls):
        cls.count += 1     # Modifies class variable
        print(f"Count: {cls.count}")

    @staticmethod
    def greet():
        print("Hello from static method!")   # No class/instance access

MyClass.increment()
MyClass.increment()
MyClass.greet()
```

```python
# Q23. How does Polymorphism work in Python with Inheritance?
# In Python, when a child class overrides a parent method, calling that method on any
# object (regardless of its type) invokes the correct version at runtime.
# This is called Runtime Polymorphism / Dynamic Dispatch.
# Python's duck typing also enables polymorphism without inheritance.

class Shape:
    def area(self):
        return 0

class Circle(Shape):
    def area(self):
        return 3.14 * 5 * 5

class Square(Shape):
    def area(self):
        return 4 * 4

shapes = [Circle(), Square(), Shape()]
for shape in shapes:
    print(f"{shape.__class__.__name__} area: {shape.area()}")   # Polymorphism in action
```

```python
# Q24. What is Method Chaining in Python OOP?
# Method Chaining is calling multiple methods on the same object in a single statement.
# Achieved by returning 'self' from each method.
# Makes code more readable and concise (fluent interface pattern).
# Common in libraries like Pandas, SQLAlchemy, Django ORM.

class Builder:
    def __init__(self):
        self.result = []

    def add(self, item):
        self.result.append(item)
        return self               # Returns self for chaining

    def build(self):
        return " -> ".join(self.result)

output = Builder().add("Step 1").add("Step 2").add("Step 3").build()
print(output)
```

```python
# Q25. What is the purpose of the __call__ method in Python?
# __call__ makes an object callable — you can call it like a function using obj().
# When obj() is written, Python internally calls obj.__call__().
# Use cases:
#   - Creating function-like objects
#   - Decorators (which are callable objects)
#   - Caching/memoization classes

class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

double = Multiplier(2)
triple = Multiplier(3)
print(double(5))   # Calls __call__(5) → 10
print(triple(5))   # Calls __call__(5) → 15
```

```python
# Practical 2: Abstract class Shape with area(), derived classes Circle and Rectangle

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius ** 2, 2)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c = Circle(7)
r = Rectangle(5, 3)
print(f"Circle area: {c.area()}")
print(f"Rectangle area: {r.area()}")
```

```python
# Practical 3: Multi-level Inheritance — Vehicle → Car → ElectricCar

class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def describe(self):
        print(f"Vehicle Type: {self.vehicle_type}")

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def describe(self):
        super().describe()
        print(f"Brand: {self.brand}")

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery = battery_capacity

    def describe(self):
        super().describe()
        print(f"Battery: {self.battery} kWh")

ev = ElectricCar("Electric", "Tesla", 100)
ev.describe()
```

```python
# Practical 4: Polymorphism with Bird base class, Sparrow and Penguin derived classes

class Bird:
    def fly(self):
        print("Some birds can fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies swiftly!")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly, they swim instead!")

birds = [Bird(), Sparrow(), Penguin()]
for bird in birds:
    bird.fly()   # Polymorphism: same method, different behavior
```

```python
# Practical 5: Encapsulation using BankAccount class

class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner = owner
        self.__balance = initial_balance   # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ₹{amount}. New balance: ₹{self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}. Remaining balance: ₹{self.__balance}")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        print(f"Account Balance for {self.owner}: ₹{self.__balance}")

acc = BankAccount("Alice", 1000)
acc.deposit(500)
acc.withdraw(200)
acc.check_balance()
acc.withdraw(5000)   # Should fail
```

```python
# Practical 9: Fraction class with __str__ override

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

    def __repr__(self):
        return f"Fraction({self.numerator}, {self.denominator})"

f = Fraction(3, 4)
print(f)         # Uses __str__
print(repr(f))   # Uses __repr__
```

```python
# Practical 10: Operator Overloading — Vector class with __add__

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):
        return self.__str__()

v1 = Vector(2, 3)
v2 = Vector(4, 1)
v3 = v1 + v2   # Calls __add__
print(f"{v1} + {v2} = {v3}")
```

```python
# Practical 12: Student class with average_grade() method

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades   # List of grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

s = Student("Priya", [85, 90, 78, 92, 88])
print(f"Student: {s.name}")
print(f"Grades: {s.grades}")
print(f"Average Grade: {s.average_grade():.2f}")
```

```python
# Practical 13: Rectangle class with set_dimensions() and area()

class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        self.length = length
        self.width = width
        print(f"Dimensions set: {self.length} x {self.width}")

    def area(self):
        return self.length * self.width

rect = Rectangle()
rect.set_dimensions(10, 5)
print(f"Area: {rect.area()}")
```

```python
# Practical 14: Employee with calculate_salary(), Manager adds bonus

class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base = super().calculate_salary()
        return base + self.bonus

emp = Employee("Amit", 160, 300)
mgr = Manager("Neha", 160, 500, 10000)

print(f"{emp.name}'s Salary: ₹{emp.calculate_salary()}")
print(f"{mgr.name}'s Salary (with bonus): ₹{mgr.calculate_salary()}")
```

```python
# Practical 16: Abstract class Animal with sound(), derived classes Cow and Sheep

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Cow says: Moo!")

class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa!")

animals = [Cow(), Sheep()]
for animal in animals:
    animal.sound()
```

```python
# Practical 17: Book class with get_book_info() method

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return (f"Title: '{self.title}' | "
                f"Author: {self.author} | "
                f"Year: {self.year_published}")

book1 = Book("The Alchemist", "Paulo Coelho", 1988)
book2 = Book("Clean Code", "Robert C. Martin", 2008)
print(book1.get_book_info())
print(book2.get_book_info())
```

```python
# Practical 18: House class and derived Mansion class with number_of_rooms attribute

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def get_info(self):
        return f"Address: {self.address} | Price: ₹{self.price:,}"

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def get_info(self):
        base_info = super().get_info()
        return f"{base_info} | Rooms: {self.number_of_rooms}"

house = House("12 MG Road, Bengaluru", 5000000)
mansion = Mansion("1 Palace Road, Bengaluru", 50000000, 20)

print("House:", house.get_info())
print("Mansion:", mansion.get_info())
```
