# Unsupervised Learning and Clustering Assignment

Theoretical questions are written as comments. Practical questions are provided as runnable Python code cells.

```python
# THEORETICAL QUESTIONS - PART 1
# Q1. What is unsupervised learning in the context of machine learning?
# Unsupervised learning is a type of machine learning where the model works with unlabeled data
# and tries to discover patterns, structures, or groupings without known target outputs.

# Q2. How does K-Means clustering algorithm work?
# K-Means chooses K initial centroids, assigns each data point to the nearest centroid,
# updates centroids as the mean of assigned points, and repeats until convergence.

# Q3. Explain the concept of a dendrogram in hierarchical clustering.
# A dendrogram is a tree-like diagram that shows how clusters are merged or split at different
# distances in hierarchical clustering.

# Q4. What is the main difference between K-Means and Hierarchical Clustering?
# K-Means partitions data into a fixed number of clusters, while hierarchical clustering builds
# a tree of nested clusters without requiring all partitions to be chosen in advance.

# Q5. What are the advantages of DBSCAN over K-Means?
# DBSCAN can detect arbitrarily shaped clusters, does not require specifying the number of clusters
# beforehand, and can identify noise points.

# Q6. When would you use Silhouette Score in clustering?
# Silhouette Score is used to evaluate how well data points fit within their assigned clusters
# and how well separated the clusters are.

# Q7. What are the limitations of Hierarchical Clustering?
# It can be computationally expensive on large datasets, sensitive to noise and outliers,
# and early merge or split decisions cannot be undone.

# Q8. Why is feature scaling important in clustering algorithms like K-Means?
# K-Means relies on distance, so features with larger numeric scales can dominate the clustering
# unless the data is scaled.

# Q9. How does DBSCAN identify noise points?
# DBSCAN labels points as noise when they do not belong to a dense region and do not have enough
# neighboring points within the eps radius.

# Q10. Define inertia in the context of K-Means.
# Inertia is the sum of squared distances between data points and their assigned cluster centroids.

# Q11. What is the elbow method in K-Means clustering?
# The elbow method plots inertia against different values of K and selects the K where the rate of
# improvement drops noticeably, forming an elbow shape.

# Q12. Describe the concept of density in DBSCAN.
# Density in DBSCAN refers to how many data points lie within a specified neighborhood radius.
# Dense regions form clusters.
```

```python
# THEORETICAL QUESTIONS - PART 2
# Q13. Can hierarchical clustering be used on categorical data?
# Yes, hierarchical clustering can be used on categorical data if a suitable distance or similarity
# measure is chosen, such as Hamming distance or Gower distance.

# Q14. What does a negative Silhouette Score indicate?
# A negative Silhouette Score indicates that a point may be assigned to the wrong cluster because it
# is closer to another cluster than to its own.

# Q15. Explain the term linkage criteria in hierarchical clustering.
# Linkage criteria define how the distance between two clusters is calculated during merging,
# such as single, complete, average, or ward linkage.

# Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
# K-Means assumes roughly spherical and similarly sized clusters, so it may split large clusters
# incorrectly or merge small dense clusters with larger sparse ones.

# Q17. What are the core parameters in DBSCAN, and how do they influence clustering?
# The core parameters are eps and min_samples. eps controls neighborhood radius, and min_samples
# controls how many points are needed to form a dense region.

# Q18. How does K-Means++ improve upon standard K-Means initialization?
# K-Means++ chooses initial centroids more carefully so they are spread out, which often leads to
# faster convergence and better clustering quality.

# Q19. What is agglomerative clustering?
# Agglomerative clustering is a bottom-up hierarchical method where each point starts as its own
# cluster and clusters are merged step by step.

# Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?
# Inertia only measures compactness, while Silhouette Score considers both compactness and separation,
# so it gives a more complete view of clustering quality.
```

## Practical Questions

```python
# Common imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.datasets import load_iris, load_wine, load_breast_cancer, load_digits
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from scipy.cluster.hierarchy import dendrogram, linkage
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 1. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200)
plt.title('K-Means on Synthetic Blobs')
plt.show()
```

```python
# 2. Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters
iris = load_iris()
agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(iris.data)

print('First 10 predicted labels:', labels[:10])
```

```python
# 3. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot
X, _ = make_moons(n_samples=300, noise=0.08, random_state=42)
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

noise_mask = labels == -1
plt.scatter(X[~noise_mask, 0], X[~noise_mask, 1], c=labels[~noise_mask], cmap='plasma', label='Clustered')
plt.scatter(X[noise_mask, 0], X[noise_mask, 1], c='red', label='Outliers')
plt.title('DBSCAN on make_moons')
plt.legend()
plt.show()
```

```python
# 4. Load the Wine dataset and apply K-Means clustering after standardizing the features
wine = load_wine()
X_scaled = StandardScaler().fit_transform(wine.data)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

unique, counts = np.unique(labels, return_counts=True)
for cluster, size in zip(unique, counts):
    print(f'Cluster {cluster}: {size} samples')
```

```python
# 5. Use make_circles to generate synthetic data and cluster it using DBSCAN
X, _ = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)
dbscan = DBSCAN(eps=0.15, min_samples=5)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm')
plt.title('DBSCAN on make_circles')
plt.show()
```

```python
# 6. Load the Breast Cancer dataset, apply MinMaxScaler, and use K-Means with 2 clusters
cancer = load_breast_cancer()
X_scaled = MinMaxScaler().fit_transform(cancer.data)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X_scaled)

print('Cluster centroids:')
print(kmeans.cluster_centers_)
```

```python
# 7. Generate synthetic data using make_blobs with varying cluster standard deviations and cluster with DBSCAN
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=[0.5, 1.5, 2.5], random_state=42)
dbscan = DBSCAN(eps=0.8, min_samples=6)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10')
plt.title('DBSCAN on Blobs with Varying Density')
plt.show()
```

```python
# 8. Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means
digits = load_digits()
X_pca = PCA(n_components=2, random_state=42).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_pca)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='tab10', s=15)
plt.title('K-Means Clusters on PCA-Reduced Digits')
plt.show()
```

```python
# 9. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []
k_values = [2, 3, 4, 5]

for k in k_values:
    labels = KMeans(n_clusters=k, random_state=42, n_init=10).fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.bar(k_values, scores, color='teal')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k Values')
plt.show()
```

```python
# 10. Load the Iris dataset and plot a dendrogram with average linkage
iris = load_iris()
linked = linkage(iris.data, method='average')

plt.figure(figsize=(10, 5))
dendrogram(linked)
plt.title('Dendrogram for Iris Dataset')
plt.show()
```

```python
# 11. Generate overlapping clusters using make_blobs, apply K-Means, and visualize decision boundaries
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='Accent')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Accent', edgecolor='k')
plt.title('K-Means with Decision Boundaries')
plt.show()
```

```python
# 12. Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE
digits = load_digits()
X_tsne = TSNE(n_components=2, random_state=42, perplexity=30).fit_transform(digits.data)
labels = DBSCAN(eps=2.5, min_samples=5).fit_predict(X_tsne)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab20', s=10)
plt.title('DBSCAN on t-SNE Reduced Digits')
plt.show()
```

```python
# 13. Generate synthetic data using make_blobs and apply Agglomerative Clustering with complete linkage
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)
agg = AgglomerativeClustering(n_clusters=3, linkage='complete')
labels = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1')
plt.title('Agglomerative Clustering with Complete Linkage')
plt.show()
```

```python
# 14. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means
cancer = load_breast_cancer()
X_scaled = StandardScaler().fit_transform(cancer.data)
k_values = range(2, 7)
inertias = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertias.append(model.inertia_)

plt.plot(list(k_values), inertias, marker='o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Inertia Values for Different K')
plt.show()
```

```python
# 15. Generate synthetic concentric circles and cluster using Agglomerative Clustering with single linkage
X, _ = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)
agg = AgglomerativeClustering(n_clusters=2, linkage='single')
labels = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm')
plt.title('Agglomerative Clustering with Single Linkage')
plt.show()
```

```python
# 16. Use the Wine dataset, apply DBSCAN after scaling the data, and count clusters excluding noise
wine = load_wine()
X_scaled = StandardScaler().fit_transform(wine.data)
labels = DBSCAN(eps=2.2, min_samples=5).fit_predict(X_scaled)

num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print('Number of clusters excluding noise:', num_clusters)
```

```python
# 27. Generate noisy blobs and use DBSCAN to identify both clusters and noise points. Print the count
X, _ = make_blobs(n_samples=350, centers=3, cluster_std=1.2, random_state=42)
noise = np.random.uniform(low=X.min(), high=X.max(), size=(30, 2))
X = np.vstack([X, noise])
labels = DBSCAN(eps=0.6, min_samples=6).fit_predict(X)

noise_count = np.sum(labels == -1)
print('Noise points detected:', noise_count)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10')
plt.title('DBSCAN with Noise Detection')
plt.show()
```
