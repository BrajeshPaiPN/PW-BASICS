## Practical Questions

```python
# Common setup
# Make sure MongoDB server is running locally or update the connection string as needed.

import pandas as pd
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['superstore_db']
orders = db['Orders']
```

```python
# 1. Load the Superstore dataset from a CSV file into MongoDB
df = pd.read_csv('superstore.csv')
orders.delete_many({})
orders.insert_many(df.to_dict(orient='records'))
print('Inserted documents:', orders.count_documents({}))
```

```python
# 2. Retrieve and print all documents from the Orders collection
for doc in orders.find():
    print(doc)
```

```python
# 3. Count and display the total number of documents in the Orders collection
total_docs = orders.count_documents({})
print('Total documents:', total_docs)
```

```python
# 4. Fetch all orders from the West region
west_orders = orders.find({'Region': 'West'})
for doc in west_orders:
    print(doc)
```

```python
# 5. Find orders where Sales is greater than 500
high_sales = orders.find({'Sales': {'$gt': 500}})
for doc in high_sales:
    print(doc)
```

```python
# 6. Fetch the top 3 orders with the highest Profit
top_profit = orders.find().sort('Profit', -1).limit(3)
for doc in top_profit:
    print(doc)
```

```python
# 7. Update all orders with Ship Mode as First Class to Premium Class
result = orders.update_many(
    {'Ship Mode': 'First Class'},
    {'$set': {'Ship Mode': 'Premium Class'}}
)
print('Matched:', result.matched_count)
print('Modified:', result.modified_count)
```

```python
# 8. Delete all orders where Sales is less than 50
result = orders.delete_many({'Sales': {'$lt': 50}})
print('Deleted documents:', result.deleted_count)
```

```python
# 9. Use aggregation to group orders by Region and calculate total sales per region
pipeline = [
    {'$group': {'_id': '$Region', 'total_sales': {'$sum': '$Sales'}}},
    {'$sort': {'total_sales': -1}}
]

for doc in orders.aggregate(pipeline):
    print(doc)
```

```python
# 10. Fetch all distinct values for Ship Mode from the collection
ship_modes = orders.distinct('Ship Mode')
print('Distinct Ship Modes:', ship_modes)
```

```python
# 11. Count the number of orders for each category
pipeline = [
    {'$group': {'_id': '$Category', 'order_count': {'$sum': 1}}},
    {'$sort': {'order_count': -1}}
]

for doc in orders.aggregate(pipeline):
    print(doc)
```
